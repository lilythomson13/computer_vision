```python
from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test
```

```python
#Q3

import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive')
Z_train=pd.read_parquet('/content/drive/MyDrive/Z_train.parquet')
Z_test=pd.read_parquet('/content/drive/MyDrive/Z_test.parquet')
```

```python
#Q4

from sklearn.neighbors import KNeighborsClassifier

k_bar=25
k_grid=np.arange(2,k_bar) #try 1,2
accuracy=np.zeros(k_bar)

for k in range(k_bar):
    print(k)
    knn=KNeighborsClassifier(n_neighbors=k+1)
    predictor=knn.fit(Z_train.values,y_train)
    accuracy[k]=predictor.score(Z_test.values,y_test)
# Bug in sklearn requires .values

accuracy_max=np.max(accuracy)
# highest recorded accuracy

max_index=np.where(accuracy==accuracy_max)

k_star=k_grid[max_index]
# Find the optimal value of k

print(k_star)

plt.plot(np.arange(0,k_bar),accuracy)
# Plot accuracy by k
plt.xlabel("k")
plt.title("optimal k:"+str(k_star))
plt.ylabel('Accuracy')
plt.show()

#Using the reshaped X_test and y_test data, the optimal number of neighbors is 4
```

```python
#Q5

knn = KNeighborsClassifier(n_neighbors=1)
predictor = knn.fit(Z_train.values,y_train)
y_hat = predictor.predict(Z_test.values)

accuracy = knn.score(Z_test.values,y_test)
print('Accuracy: ', accuracy)

pd.crosstab(y_test, y_hat)
```

```python
#The knn predictor performs very well on the test set
#Using a k value of 4, there is almost 97% accuracy
#I can see in the confusion matrix that there are some numbers that get confused, but overall there is very high accuracy
```
